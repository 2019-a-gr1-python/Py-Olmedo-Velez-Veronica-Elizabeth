# Marathon de Boston 2019 Resultados
Dataframe con los resultados de la marathon realizada en Boston en el 2019. Incluye la tabla de posiciones y los tiempos promedios para las distintas distancias

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
```

# Tabla de participacion por edad para Mujeres

```python
data = pd.read_csv('data/marathon_boston_2019.csv')
fig = plt.figure(figsize=(8,18))
data.Age[data.Gender=='F'].value_counts().plot(kind='barh',alpha=1)
plt.show()
```

```python
try:
    with open('./data/marathon_boston_2019.csv')as archivo_abierto:
        genero = []
        edad = []
        tiempo = []
        arreglo_lineas_archivo = archivo_abierto.readlines()
        for linea in arreglo_lineas_archivo:
            data = linea.split(',')
            genero.append(data[4])
            edad.append(data[3])
            tiempo.append(data[17])
        genero.remove('Gender')
        edad.remove('Age')
        tiempo.remove('OfficialTime')
            
except Exception:
    print("No se pudo leer el archivo")
```

# Tabla de Participacion por genero

```python
generoC = Counter(genero).values()
letra_genero = ['Masculino','Femenino']

plt.figure(figsize=(5,6))
plt.title('Participacion por genero')
plt.bar(letra_genero, generoC, color = 'blue')
for a,b in zip(letra_genero, generoC):
    plt.text(a,b,str(b), ha='center',fontweight = 'bold')
plt.show()

plt.pie(generoC, labels=letra_genero, autopct='%1.1f%%',
        shadow=True, startangle=90)
plt.axis('equal')

plt.show()
```

```python
# Grafico de personas jovenes que participarón

edadNum = [int(x) for x in edad]
a = 0 
b = 0
f = 0
for i in edadNum:
    if(i<=25):
        a = a+1
    else:
        f = f+1

edades = [a,f]
edadesCat = ['De 18 a 25 años', 'De 26 años en adelante']


plt.pie(edades, labels=edadesCat, autopct='%1.1f%%',
        shadow=True, startangle=90, radius = 1800)
plt.axis('equal')
plt.show()
```

# Tabla de participacion por genero y por edad

```python
edadNum = [int(x) for x in edad]
aM = 0
aF = 0
bM = 0 
bF = 0 
cM = 0
cF = 0
dM = 0 
dF = 0 
eM = 0
eF = 0
fM = 0
fF = 0

for i,j in zip(edadNum,genero):
    if(i<=25 and j=='M'):
        aM = aM+1
    elif(i<=25 and j=='F'):
        aF = aF+1
    elif(i<=35 and i>25 and j=='M'):
        bM = bM+1
    elif(i<=35 and i>25 and j=='F'):
        bF = bF+1
    elif(i<=45 and i>35 and j=='M'):
        cM = cM+1
    elif(i<=45 and i>35 and j=='F'):
        cF = cF+1
    elif(i<=55 and i>45 and j=='M'):
        dM = dM+1
    elif(i<=55 and i>45 and j=='F'):
        dF = dF+1
    elif(i<=65 and i>55 and j=='M'):
        eM = eM+1
    elif(i<=65 and i>55 and j=='F'):
        eF = eF+1
    elif(i>65 and j=='M'):
        fM = fM+1
    else:
        fF = fF+1
        
edadesM = [aM,bM,cM,dM,eM,fM]
edadesF = [aF,bF,cF,dF,eF,fF]
edadesCat = ['De 18 a 25 años', 'De 26 a 35 años', 'De 36 a 45 años', 
             'De 46 a 55 años', 'De 56 a 65 años', 'De 65 años en adelante']
N = 6
ind = np.arange(N)
width = 0.35 
fig, ax = plt.subplots(figsize=(15,8))
bar1 = ax.bar(ind, edadesM, width, color = 'blue')
for a,b in zip(ind, edadesM):
    ax.text(a,b,str(b), ha='center',fontweight = 'bold')
bar2 = ax.bar(ind + width, edadesF, width, color = 'red')
for a,b in zip(ind + width, edadesF):
    ax.text(a,b,str(b), ha='center',fontweight = 'bold')

ax.set_title('Participacion por Genero y por Edad')
ax.set_xticks(ind + width / 2)
ax.set_xticklabels(('De 18 a 25 años', 'De 26 a 35 años', 'De 36 a 45 años', 
             'De 46 a 55 años', 'De 56 a 65 años', 'De 65 años en adelante'))
ax.legend((bar1[0], bar2[0]), ('Masculino', 'Femenino'))
plt.show()
```
