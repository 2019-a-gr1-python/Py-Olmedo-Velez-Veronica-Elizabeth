# Operaciones 

```python
def calculadora(array1, array2, operacion='Sumar'):
    
    def sumar():
        return array1 + array2
    
    def restar():
        return array1 - array2
    
    def dividir():
        return array1 / array2
    
    def multiplicar():
        return array1 * array2
    
    def switch_operacion():
        return {
            'Sumar': sumar(),
            'Restar': restar(),
            'Multiplicar': multiplicar(),
            'Dividir': dividir()
        }[operacion]
    
    return switch_operacion() 
```

```python
calculadora(A,B, 'Restar') 
```

```python
calculadora(A,B, 'Sumar') 
```

```python
calculadora(A,B, 'Multiplicar') 
```

```python
calculadora(A,5, 'Sumar') 
```

```python
calculadora(A,5, 'Restar') 
```

```python
calculadora(A,5, 'Multiplicar') 
```

```python
calculadora(A,5, 'Dividir') 
```

# Comprobar si una matriz es inversa 

```python
# Matriz 2x2
B = np.array([[2.,1.],[5.,3.]])
print (B)
B.shape
```

```python
# Inversa de B
B_inv = np.linalg.inv(B) 
print("B inversa")
print(B_inv) 
```

```python
# Determinante 
B_det = np.linalg.det(B)
print(B_det) 
```

```python
# Round redondea una matriz al número dado de decimales.
if(round(1/B_det)!=0):
    print("Es invertible / Regular")
else:
    print("Matriz Singular") 
```

# Generar un rompecabezas

```python
# Cara de un mapache
mapache = misc.face()  
def get_puzzle_imagen(numero_partes):
    mapache_partes = np.vsplit(mapache,numero_partes)
    piezas_indices = np.random.permutation(numero_partes)
    mapache_mezclado = np.concatenate(list(map(lambda x: mapache_partes[x],piezas_indices)))
    plt.imshow(mapache_mezclado)
    plt.show()
```

```python
get_puzzle_imagen(32)
get_puzzle_imagen(4)
```

# Cifrar imagen

```python
mapache = misc.face()
def cifrar_imagen(key,imagen):
    return np.array([list(map(lambda e:(e*key)%300,x)) for x in imagen])

plt.imshow(cifrar_imagen(100,mapache))
plt.show()
```
